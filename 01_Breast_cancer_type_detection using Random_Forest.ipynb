```python
import pandas as pd
bre_cancer = pd.read_csv("breast_cancer.csv")
```

```python
bre_cancer.shape
```

```python
bre_cancer.info()
```

```python
bre_cancer.head()
```

```python
bre_cancer.diagnosis.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
bre_cancer['diagnosis']= le.fit_transform(bre_cancer.diagnosis)
bre_cancer
```

```python
bre_cancer.head()
```

```python
bre_cancer.columns
```

```python
from sklearn.model_selection import train_test_split
x = bre_cancer.iloc[:,2:31].values #features
y = bre_cancer.diagnosis #labels
```

```python
x.shape
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
# Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=40)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
#Import metrix module for accuracy calculation
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test,y_pred))
```

```python
clf.feature_importances_
```

```python
from sklearn.metrics import classification_report, confusion_matrix,\
 accuracy_score

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test, y_pred))
```

# Using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
y_pred
```

```python
print(classification_report(y_test, y_pred))
print('\nAccuracy: {0:.2f}'.format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fn =  bre_cancer.iloc[:,2:31].values #features
cn = ['diagnosis']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (2,3), dpi=300)
tree.plot_tree(clf,
               feature_names = fn,
               class_names=cn,
               filled = True);
fig.savefig('classification_tree_bre_cancer.png')
```
